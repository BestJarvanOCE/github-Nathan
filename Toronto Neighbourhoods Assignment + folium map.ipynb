```python
import pandas as pd
import numpy as np
```

Importing pandas and reading the data file turned from a wikipedia page to a csv.

```python
df = pd.read_csv('/resources/data/Neighbourhoods_in_Toronto.csv')
df.head()
```

Dropping rows that have "Not Assigned" as boroughs

```python
df = df[df.Borough != 'Not assigned']
df.head()
```

Combining neighbourhoods that share the same boroughs.

```python
df = df.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)
df
```

Boroughs with no assigned neighbourhood name are granted its borough's name.

```python
df['Neighbourhood']=np.where(
    df['Neighbourhood'] == 'Not assigned',
    df['Borough'], df['Neighbourhood'])
df.head(12)
```

```python
PC_Loc = pd.read_csv('http://cocl.us/Geospatial_data')
PC_Loc.head(10)
```

Changing the column name Postal Code to Postcode for diction's sake.

```python
PC_Loc.rename(columns={'Postal Code':'Postcode'}, inplace = True)
PC_Loc.head(10)
```

Merging the two tables together by the similar variable 'Postcode'

```python
df_loc = pd.merge(PC_Loc, df, on = 'Postcode')
df_loc.head(10)
```

The new table (df_loc_tor) used for plotting has its data points with boroughs within Toronto.

```python
df_loc_tor = df_loc[df_loc['Borough'].str.contains('Toronto',regex=False)]
df_loc_tor.head(10)
```

Creation of folium map

```python
import folium
```

```python
Toronto_map = folium.Map(location=[43.651,-79.347],zoom_start=12)
for lat,lng,borough,neighbourhood in zip(df_loc_tor['Latitude'],df_loc_tor['Longitude'],df_loc_tor['Borough'],df_loc_tor['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=6,
    popup=label,
    color='red',
    fill=True,
    fill_color='#db4242',
    fill_opacity=0.8,
    parse_html=False).add_to(Toronto_map)
Toronto_map
```
